```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
add_gift = pd.read_parquet('/Users/colin/Downloads/gift-add') \
    .query("date>'2023-06-09 02:00:00' & index==0 & (name.str.contains('\?') == False)")
add_gift
```

```python

# æ˜ å°„
price_dict = {
        'ç³–æœ': 1,
        'ç”Ÿæ—¥è›‹ç³•': 52,
        'ç³–æœé£èˆ¹': 52,
        'æ¢¦å¹»æ³¡æ²«': 52,
        'ä»²å¤ä¹‹æ¢¦': 99,
        'è¤ç«ä¹‹å…‰': 99,
        'æ·±æµ·æ¢¦å¢ƒ': 188,
        'æ˜Ÿæ„¿æµæ˜Ÿ': 288,
        'è¤ç«ä¹‹æ‹': 520,
        'èµ›åšä¸–ç•Œ': 999,
}

add_gift['price'] = add_gift['content'].map(price_dict)
add_gift['total'] = add_gift['price'] * add_gift['num']

add_gift
```

```python
# é€‰æ‹©ç”¨æˆ·
id_data = add_gift.reset_index()

# çœç•¥é«˜ç²¾åº¦æ•°æ®
id_data['date'] = id_data['date'].dt.round('D')
id_data = id_data.drop(columns=['index'])
id_data = id_data.drop_duplicates()

# æ±‡æ€»
id_data
```

## å¯è§†åŒ–

```python
# Group by salesperson and sum the amounts
sales_summary = id_data.groupby('name')['total'].sum().sort_values(ascending=False)

# Create the bar chart
sales_summary.plot(kind='bar')

plt.title('Sales by Salesperson')
plt.ylabel('Amount')
plt.show()
```
