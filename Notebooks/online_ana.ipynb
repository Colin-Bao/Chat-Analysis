```python
# ÈÄâÊã©Áî®Êà∑
id_data = user_data.query("Name=='Â∞ΩÊ¨¢'").reset_index()

# ÁúÅÁï•È´òÁ≤æÂ∫¶Êï∞ÊçÆ
# id_data['Date'] = id_data['Date'].dt.round('min')

# ÂéªÈáç
# id_data = id_data.drop_duplicates(['Date'])

id_data
```

## Êï∞ÊçÆÈáçÈááÊ†∑
ÂàõÂª∫‰∏Ä‰∏™ÂåÖÂê´‰∏ÄÂ§©ÊØèÂàÜÈíüÊó∂Èó¥Èó¥ÈöîÁöÑÊó∂Èó¥ËåÉÂõ¥ÔºåÂπ∂Â∞ÜÂÖ∂‰Ωú‰∏∫DataFrameÁöÑÁ¥¢Âºï„ÄÇ‰Ω†ÂèØ‰ª•‰ΩøÁî®pd.date_rangeÂáΩÊï∞Êù•ÂàõÂª∫Êó∂Èó¥ËåÉÂõ¥„ÄÇ

```python
# ÊåâËßÑÂàôÈáçÊñ∞ÈááÊ†∑
df_res = id_data.set_index('Date').resample('2T').first()
df_res['Name'] = df_res['Name'].fillna(method='ffill')
df_res['Online'] = df_res['Online'].fillna(0)
# df_res['Online'] = df_res['Online'] * df_res['Text']

# Êú™ÈááÊ†∑Âà∞ÁöÑÊï∞ÊçÆ
df_res['cover'] = df_res['index'].apply(lambda x: -1 if np.isnan(x) else 0)
df_res
```

## ÂèØËßÜÂåñ

```python
fig, ax = plt.subplots(figsize=(10, 5))

# ÂàõÂª∫Êü±Áä∂Âõæ
ax.bar(df_res.index, df_res['Online'], width=0.001, align='edge', color='skyblue', alpha=0.9)
# ax.bar(df_res.index, df_res['cover'], width=0.001, align='edge', color='b', alpha=0.2)

# ËÆæÁΩÆÊó•ÊúüÊ†ºÂºè
ax.xaxis_date()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))

# ÂàõÂª∫‰∏Ä‰∏™ÂÖ±‰∫´xËΩ¥ÁöÑÁ¨¨‰∫å‰∏™ËΩ¥
ax2 = ax.twinx()

# Âú®Á¨¨‰∫å‰∏™ËΩ¥‰∏äÁªòÂà∂ÊäòÁ∫øÂõæ
ax2.plot(df_res.index, df_res['index'], 'r-')
ax2.set_ylabel('Rank')
ax2.invert_yaxis()  # ÈÄÜËΩ¨Á¨¨‰∫å‰∏™yËΩ¥ÁöÑÈ°∫Â∫è

# ËÆæÁΩÆÊ†áÈ¢òÂíåÊ†áÁ≠æ
ax.set_xlabel('Time')
ax.set_ylabel('Online Status')
ax.set_title('Online Status of User')

# Ê∑ªÂä†ÁΩëÊ†º
ax.grid(False)

# Ëá™Âä®ÊóãËΩ¨Êó•ÊúüÊ†áÁ≠æ‰ª•ÈÅøÂÖçÈáçÂè†
# fig.autofmt_xdate()

plt.tight_layout()
plt.show()
```
