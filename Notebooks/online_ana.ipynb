## 个体筛选

```python
# 选择用户
id_data = user_data.query("Name=='尽欢'").reset_index()

# 省略高精度数据
# id_data['Date'] = id_data['Date'].dt.round('min')

# 去重
# id_data = id_data.drop_duplicates(['Date'])

id_data
```

## 数据重采样
创建一个包含一天每分钟时间间隔的时间范围，并将其作为DataFrame的索引。你可以使用pd.date_range函数来创建时间范围。

```python
# 按规则重新采样
df_res = id_data.set_index('Date').resample('2T').first()
df_res['Name'] = df_res['Name'].fillna(method='ffill')
df_res['Online'] = df_res['Online'].fillna(0)
# df_res['Online'] = df_res['Online'] * df_res['Text']

# 未采样到的数据
df_res['cover'] = df_res['index'].apply(lambda x: -1 if np.isnan(x) else 0)
df_res
```

## 可视化

```python
fig, ax = plt.subplots(figsize=(10, 5))

# 创建柱状图
ax.bar(df_res.index, df_res['Online'], width=0.001, align='edge', color='skyblue', alpha=0.9)
# ax.bar(df_res.index, df_res['cover'], width=0.001, align='edge', color='b', alpha=0.2)

# 设置日期格式
ax.xaxis_date()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))

# 创建一个共享x轴的第二个轴
ax2 = ax.twinx()

# 在第二个轴上绘制折线图
ax2.plot(df_res.index, df_res['index'], 'r-')
ax2.set_ylabel('Rank')
ax2.invert_yaxis()  # 逆转第二个y轴的顺序

# 设置标题和标签
ax.set_xlabel('Time')
ax.set_ylabel('Online Status')
ax.set_title('Online Status of User')

# 添加网格
ax.grid(False)

# 自动旋转日期标签以避免重叠
# fig.autofmt_xdate()

plt.tight_layout()
plt.show()
```
